```
patient_allergies = pd.merge(allergies, patients, on=['PATIENT'], how='right')
patient_allergies.drop(0, inplace=True)

patient_observations = pd.merge(observations, patients, on=['PATIENT'], how='right')
patient_observations.drop(0, inplace = True)

patient_conditions = pd.merge(conditions, patients, on=['PATIENT'], how='right')
patient_conditions.drop(0, inplace=True)

patient_payer_transitions = pd.merge(payer_transitions, patients, on=['PATIENT'], how='right')
patient_payer_transitions.drop(0, inplace=True)

payers_p_transitions = pd.merge(payer_transitions, payers, on=['PAYER'], how='right')
payers_p_transitions.drop(0, inplace=True)

patient_medications = pd.merge(medications, patients, on=['PATIENT'], how= 'right')
patient_medications.drop(0, inplace=True)

patient_encounters = pd.merge(encounters, patients, on=['PATIENT'], how='right')
patient_encounters.drop(0, inplace=True)
```

```
grouped_patients = required_data.groupby('PATIENT')

for name, group in grouped_patients:
  data_dict = {}
  data_dict['PATIENT'] = name
  for ind in group.index:
    data_dict['DATE'] = group['obs_DATE'][ind]
    data_dict['ENCOUNTER'] = group['ENCOUNTER'][ind]
    data_dict['obs_CODE'] = group['obs_CODE'][ind]
    data_dict['PREFIX'] = group['PREFIX'][ind]
    data_dict['FIRST'] = group['FIRST'][ind]
    data_dict['LAST'] = group['LAST'][ind]
    data_dict['GENDER'] = group['GENDER'][ind]
    data_dict['BIRTHDATE'] = group['BIRTHDATE'][ind]
    data_dict['DEATHDATE'] = group['DEATHDATE'][ind]
    data_dict['HEALTHCARE_EXPENSES'] = group['HEALTHCARE_EXPENSES'][ind]
    data_dict['HEALTHCARE_COVERAGE'] = group['HEALTHCARE_COVERAGE'][ind]
    #print(group['obs_DESCRIPTION'][ind] + "  " + units[group['obs_DESCRIPTION'][ind]])
    data_dict[group['obs_DESCRIPTION'][ind] + "  " + units[group['obs_DESCRIPTION'][ind]]] = group['VALUE'][ind]
  df_patient_observations = df_patient_observations.append(data_dict, ignore_index=True)
df_patient_observations
```

```
patient_payers = payers_p_transitions[['PATIENT', 'START_YEAR', 'END_YEAR', 'PAYER', 'OWNERSHIP', 'payer_NAME', 'payer_CITY', 'STATE_HEADQUARTERED',
                                       'AMOUNT_COVERED', 'AMOUNT_UNCOVERED', 'payer_REVENUE',
                                       'COVERED_ENCOUNTERS', 'UNCOVERED_ENCOUNTERS', 'COVERED_MEDICATIONS',
                                       'UNCOVERED_MEDICATIONS', 'COVERED_PROCEDURES', 'UNCOVERED_PROCEDURES',
                                       'COVERED_IMMUNIZATIONS', 'UNCOVERED_IMMUNIZATIONS']] 
# patient_payers

cols = ['PATIENT', 'INSURANCE_START_YEAR', 'INSURANCE_END_YEAR', 'PAYER', 'OWNERSHIP', 'payer_NAME',
       'payer_CITY', 'STATE_HEADQUARTERED', 'AMOUNT_COVERED',
       'AMOUNT_UNCOVERED', 'payer_REVENUE', 'COVERED_ENCOUNTERS',
       'UNCOVERED_ENCOUNTERS', 'COVERED_MEDICATIONS', 'UNCOVERED_MEDICATIONS',
       'COVERED_PROCEDURES', 'UNCOVERED_PROCEDURES', 'COVERED_IMMUNIZATIONS',
       'UNCOVERED_IMMUNIZATIONS']

df_patient_payer = pd.DataFrame(columns = cols)
df_patient_payer
```

```
grouped_patients = patient_payers.groupby('PATIENT')
for name, group in grouped_patients:
  data_dict = {'PAYER' : [], 'OWNERSHIP':[], 'payer_NAME':[], 'payer_CITY':[], 'STATE_HEADQUARTERED':[],
               'AMOUNT_COVERED':[], 'AMOUNT_UNCOVERED':[], 'payer_REVENUE':[], 'COVERED_ENCOUNTERS':[],
               'UNCOVERED_ENCOUNTERS':[], 'COVERED_MEDICATIONS':[], 'UNCOVERED_MEDICATIONS':[],
               'COVERED_PROCEDURES': [], 'COVERED_IMMUNIZATIONS':[], 'UNCOVERED_IMMUNIZATIONS':[]}
  data_dict['PATIENT'] = name
  for ind in group.index:
    data_dict['INSURANCE_START_YEAR'] = group['START_YEAR'][ind]
    data_dict['INSURANCE_END_YEAR'] = group['END_YEAR'][ind]
    data_dict['PAYER'].append(group['PAYER'][ind])
    data_dict['OWNERSHIP'].append(group['OWNERSHIP'][ind])
    data_dict['payer_NAME'].append(group['payer_NAME'][ind])
    data_dict['payer_CITY'].append(group['payer_CITY'][ind])
    data_dict['STATE_HEADQUARTERED'].append(group['STATE_HEADQUARTERED'][ind])
    data_dict['AMOUNT_COVERED'].append(group['AMOUNT_COVERED'][ind])
    data_dict['AMOUNT_UNCOVERED'].append(group['AMOUNT_UNCOVERED'][ind])
    data_dict['payer_REVENUE'].append(group['payer_REVENUE'][ind])
    data_dict['COVERED_ENCOUNTERS'].append(group['COVERED_ENCOUNTERS'][ind])
    data_dict['UNCOVERED_ENCOUNTERS'].append(group['UNCOVERED_ENCOUNTERS'][ind])
    data_dict['COVERED_MEDICATIONS'].append(group['COVERED_MEDICATIONS'][ind])
    data_dict['UNCOVERED_MEDICATIONS'].append(group['UNCOVERED_MEDICATIONS'][ind])
    data_dict['COVERED_PROCEDURES'].append(group['COVERED_PROCEDURES'][ind])
    data_dict['COVERED_IMMUNIZATIONS'].append(group['COVERED_IMMUNIZATIONS'][ind])
    data_dict['UNCOVERED_IMMUNIZATIONS'].append(group['UNCOVERED_IMMUNIZATIONS'][ind])
  df_patient_payer = df_patient_payer.append(data_dict, ignore_index=True)
df_patient_payer
```

```
df_patient_allergies = pd.DataFrame(columns = ['PATIENT', 'allergies_START_DATE', 'allergies_STOP_DATE', 'CODE','allergies_DESCRIPTION'])
df_patient_allergies
```

```
grouped_patients = patient_allergies.groupby('PATIENT')
for name, group in grouped_patients:
  data_dict = {'allergies_START_DATE': [], 'allergies_STOP_DATE': [], 'CODE':[], 'allergies_DESCRIPTION':[]}
  data_dict['PATIENT'] = name
  for ind in group.index:
    data_dict['allergies_START_DATE'].append(group['allergies_START'][ind])
    data_dict['allergies_STOP_DATE'].append(group['allergies_STOP'][ind])
    data_dict['CODE'].append(group['CODE'][ind])
    data_dict['allergies_DESCRIPTION'].append(group['allergies_DESCRIPTION'][ind])
  df_patient_allergies = df_patient_allergies.append(data_dict, ignore_index=True)
df_patient_allergies
```
