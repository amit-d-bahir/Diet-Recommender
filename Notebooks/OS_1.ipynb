```
patients_data = pd.merge(observations, patients, on=['PATIENT'], how='right')

factors = ['Body Height', 'Body Weight', 'Body Mass Index', 'Diastolic Blood Pressure', 'Systolic Blood Pressure', 'Heart rate', 'Respiratory rate',
       'Hemoglobin [Mass/volume] in Blood', 'Body temperature', 'Glucose', 'Total Cholesterol', 'Low Density Lipoprotein Cholesterol', 'High Density Lipoprotein Cholesterol',
       'Are you covered by health insurance or some other kind of health care plan [PhenX]','Tobacco smoking status NHIS',
       'QALY', 'DALY', 'QOLS']
req_data = patients_data[['obs_DATE','PATIENT','ENCOUNTER', 'obs_CODE','PREFIX', 'FIRST', 'LAST', 'GENDER', 'BIRTHDATE', 'DEATHDATE','obs_DESCRIPTION', 'VALUE', 'UNITS']][patients_data['obs_DESCRIPTION'].isin(factors)]


cols = ['DATE', 'PATIENT', 'ENCOUNTER', 'obs_CODE', 'PREFIX', 'FIRST', 'LAST', 'GENDER', 'BIRTHDATE', 'DEATHDATE', 'Body Height','Body Height UNIT', 
       'Body Weight', 'Body Weight UNIT', 'Body Mass Index', 'Body Mass Index UNIT',
       'Diastolic Blood Pressure', 'Diastolic Blood Pressure UNIT', 'Systolic Blood Pressure','Systolic Blood Pressure UNIT',
       'Heart rate', 'Heart rate UNIT',  'Respiratory rate', 'Respiratory rate UNIT', 'Hemoglobin [Mass/volume] in Blood', 'Hemoglobin [Mass/volume] in Blood UNIT', 
       'Body temperature', 'Body temperature UNIT','Glucose', 'Glucose UNIT', 'Total Cholesterol','Total Cholesterol UNIT',
       'Low Density Lipoprotein Cholesterol', 'Low Density Lipoprotein Cholesterol UNIT',
       'High Density Lipoprotein Cholesterol', 'High Density Lipoprotein Cholesterol UNIT', 
       'Are you covered by health insurance or some other kind of health care plan [PhenX]',
       'Tobacco smoking status NHIS', 'Tobacco smoking status NHIS UNIT',
       'QALY', 'QALY UNIT', 'DALY','DALY UNIT' ,'QOLS', 'QOLS UNIT']
df = pd.DataFrame(columns=cols)
df
```

```
patients_group = req_data.groupby('PATIENT')

for name, group in patients_group:
  data_dict = {}
  data_dict['PATIENT'] = name 
  for ind in group.index:
    data_dict['DATE'] = group['obs_DATE'][ind]
    data_dict['ENCOUNTER'] = group['ENCOUNTER'][ind]
    data_dict['obs_CODE'] = group['obs_CODE'][ind]
    data_dict['PREFIX'] = group['PREFIX'][ind]
    data_dict['FIRST'] = group['FIRST'][ind]
    data_dict['LAST'] = group['LAST'][ind]
    data_dict['GENDER'] = group['GENDER'][ind]
    data_dict['BIRTHDATE'] = group['BIRTHDATE'][ind]
    data_dict['DEATHDATE'] = group['DEATHDATE'][ind]
    data_dict[group['obs_DESCRIPTION'][ind]] = group['VALUE'][ind]
    unit_name = group['obs_DESCRIPTION'][ind] + " UNIT" 
    data_dict[unit_name] = group['UNITS'][ind]
  df = df.append(data_dict, ignore_index=True)
```
