```
required_data = patient_observations[['obs_DATE','PATIENT','ENCOUNTER', 'obs_CODE','PREFIX', 'FIRST', 'LAST', 'GENDER', 'BIRTHDATE', 'DEATHDATE','obs_DESCRIPTION', 'VALUE', 'UNITS']][patient_observations['obs_DESCRIPTION'].isin(factors)]
required_data
```

```
units = {}
for ind in required_data.index:
  units[required_data['obs_DESCRIPTION'][ind]] = required_data['UNITS'][ind]
units
```

```
cols = ['DATE', 'PATIENT', 'ENCOUNTER', 'obs_CODE', 'PREFIX', 'FIRST', 'LAST', 'GENDER', 'BIRTHDATE', 'DEATHDATE',
        'Body Height' + "  " + units['Body Height'],  'Body Weight' + "  " + units['Body Weight'],
        'Body Mass Index' + "  " + units['Body Mass Index'], 'Body temperature' + "  " + units['Body temperature'],
        'Calcium'+"  "+units['Calcium'], 'Carbon Dioxide'+"  "+units['Carbon Dioxide'], 
        'Creatinine'+"  "+units['Creatinine'],  'Potassium'+"  "+units['Potassium'],  'Sodium'+"  "+units['Sodium'],
        'Heart rate'+"  "+units['Heart rate'],  'Respiratory rate'+"  "+units['Respiratory rate'],
        'Total Cholesterol'+"  "+units['Total Cholesterol'],
        'Low Density Lipoprotein Cholesterol'+"  "+units['Low Density Lipoprotein Cholesterol'],
        'Protein [Mass/volume] in Serum or Plasma'+"  "+units['Protein [Mass/volume] in Serum or Plasma'],
        'Protein [Mass/volume] in Urine by Test strip'+"  "+units['Protein [Mass/volume] in Urine by Test strip'],
        'Systolic Blood Pressure'+"  "+units['Systolic Blood Pressure'],
        'Diastolic Blood Pressure'+"  "+units['Diastolic Blood Pressure'],
        'Glucose [Mass/volume] in Urine by Test strip'+"  "+units['Glucose [Mass/volume] in Urine by Test strip'],
        'Hemoglobin [Mass/volume] in Blood'+"  "+units['Hemoglobin [Mass/volume] in Blood'],
        'Hematocrit [Volume Fraction] of Blood by Automated count'+"  "+units['Hematocrit [Volume Fraction] of Blood by Automated count'],
        'MCH [Entitic mass] by Automated count'+"  "+units['MCH [Entitic mass] by Automated count'],
        'MCHC [Mass/volume] by Automated count'+"  "+units['MCHC [Mass/volume] by Automated count'],
        'MCV [Entitic volume] by Automated count'+"  "+units['MCV [Entitic volume] by Automated count'],
        'Leukocytes [#/volume] in Blood by Automated count'+"  "+units['Leukocytes [#/volume] in Blood by Automated count'],
        'Erythrocytes [#/volume] in Blood by Automated count'+"  "+units['Erythrocytes [#/volume] in Blood by Automated count'],
        'Globulin [Mass/volume] in Serum by calculation'+"  "+units['Globulin [Mass/volume] in Serum by calculation'],
        'Albumin [Mass/volume] in Serum or Plasma' + "  " + units['Albumin [Mass/volume] in Serum or Plasma'],
        'Bilirubin.total [Mass/volume] in Serum or Plasma' +"  " + units['Bilirubin.total [Mass/volume] in Serum or Plasma'],
        'pH of Urine by Test strip'+"  "+units['pH of Urine by Test strip']
        ]

patient_df = pd.DataFrame(columns=cols)
patient_df
```

```
grouped_patients = required_data.groupby('PATIENT')

for name, group in grouped_patients:
  data_dict = {}
  data_dict['PATIENT'] = name
  for ind in group.index:
    data_dict['DATE'] = group['obs_DATE'][ind]
    data_dict['ENCOUNTER'] = group['ENCOUNTER'][ind]
    data_dict['obs_CODE'] = group['obs_CODE'][ind]
    data_dict['PREFIX'] = group['PREFIX'][ind]
    data_dict['FIRST'] = group['FIRST'][ind]
    data_dict['LAST'] = group['LAST'][ind]
    data_dict['GENDER'] = group['GENDER'][ind]
    data_dict['BIRTHDATE'] = group['BIRTHDATE'][ind]
    data_dict['DEATHDATE'] = group['DEATHDATE'][ind]
    #print(group['obs_DESCRIPTION'][ind] + "  " + units[group['obs_DESCRIPTION'][ind]])
    data_dict[group['obs_DESCRIPTION'][ind] + "  " + units[group['obs_DESCRIPTION'][ind]]] = group['VALUE'][ind]
  patient_df = patient_df.append(data_dict, ignore_index=True)
patient_df
```
